casos subamortiguado, sobreamortiguado y amortiguamiento critico del SLIT para $x(t)= u(t)$


```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

T = 2 #3 segundos de grafica
Fs = 1000 # 500 muestras por segundo Ajustar según señal estudiada
Nm = T*Fs #numero de muestras requeridas
tv = np.linspace(0,T/2,Nm)

L=8*(10**(-7))
C=0.1*(10**(-7))
R=15
k=1
wn= (1/L*C)**(1/2)
zeta= (R/2)*((C/L)**(1/2))
wd= (1/L*C *(1-(R**2)/4)*C/L)**(1/2)
#Caso 1 subamortiguado 0<zeta<1
yt= 1- (np.exp(-zeta*wn*tv)/(1-zeta**2)**(1/2))*np.sin(wd*tv+ np.arctan((1-zeta**2)**(1/2)/zeta))


plt.plot(tv,yt,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("y(t)",fontsize = 14)
```

```
#Caso 2 amortiguamiento critico zeta=1
L=10*(10**(-7))
C=0.4*(10**(-7))
R=10
k=1
wn= (1/L*C)**(1/2)
zeta= (R/2)*((C/L)**(1/2))
wd= (1/L*C *(1-(R**2)/4)*C/L)**(1/2)

yt= 1- np.exp(-wn*tv)*(1+wn*tv)

plt.plot(tv,yt,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("y(t)",fontsize = 14)
```

```
#Caso 3 subamortiguado zeta>1
L=15*(10**(-7))
C=2*(10**(-7))
R=10
k=1
wn= (1/L*C)**(1/2)
zeta= (R/2)*((C/L)**(1/2))
wd= (1/L*C *(1-(R**2)/4)*C/L)**(1/2)

yt= 1+ (1/2*((zeta**2 -1)**(1/2))*(zeta+(zeta**2 -1)**(1/2)) )*np.exp(-(zeta+(zeta**2 -1)**(1/2))*wn*tv)- (1/2*((zeta**2 -1)**(1/2))*(zeta-(zeta**2 -1)**(1/2)) )*np.exp(-(zeta+(zeta**2 -1)**(1/2))*wn*tv)

plt.plot(tv,yt,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("y(t)",fontsize = 14)
```
