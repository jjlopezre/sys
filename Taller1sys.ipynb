$y[n] = \sum^n_{k=-\infty}{x[k]}$.


```python
N = 7

y1= np.zeros(N)
y2= np.zeros(N)
y3= np.zeros(N)
y4= np.zeros(N)
salidaSistema2= np.zeros(N)
a = 4
b = 8
suma1=0
suma2=0
acumulador=0
x1 = np.arange(0,7,1)
print("x1", x1)
x2 = np.arange(0,14,2)
print("x2", x2)

def calcularSalidas(escalar, entrada):
  ytemp = np.zeros(N)
  for n in range(0, N):
    suma1= acumulador + entrada[n]
    ytemp[n] = escalar*suma1
  return ytemp

y1 = calcularSalidas(a, x1)
y2 = calcularSalidas(b, x2)


salidaSistema1 = y1 + y2
print('salida1 = ', salidaSistema1)


for n in range(0, N):
  y3[n] = x1[n]*a
  y4[n] = x2[n]*b

sumaEntradas = y3+y4

for n in range(0, N):
  salidaSistema2[n] = suma2 + sumaEntradas[n]


print('salida2', salidaSistema2)

for n in range(0, N):
  print('iguales', math.ceil(salidaSistema1[n]) == math.ceil(salidaSistema2[n]))
 
#grafica seal 
N = 15
n = np.linspace(-N,N,N)
ytemp = np.zeros(N)

for i in range(0, N):
    ytemp[i] = acumulador + n[i]
y = ytemp

plt.stem(n,y,label="$y[n] = \sum^n_{k=-\infty}{x[k]}$")
plt.grid()
plt.legend(loc='upper left')
plt.xlim(-N-1,N+1)
plt.xlabel("n")
plt.ylabel("$y[n]$")
```

```python
#Funcion Profesor
%matplotlib inline
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
#crear seales
h = np.r_[np.ones((5,1)),np.zeros((5,1)),np.ones((10,1))]
x = signal.triang(12).reshape(-1,1)

#funcion convolucion
def convgraf(h,x):
   lx = len(x)
   lh = len(h)
   M = 2*lx+lh
   xm = np.zeros((M,M))
   hm = np.r_[np.zeros((lx,1)),h,np.zeros((lx,1))]
   ym = np.zeros((M,M))
   for i in range(M-lx+1):
     #print(i)
     xm[i:i+lx,i] = np.flip(x).reshape(-1)
     ym[i,i] = xm[:,i].T.dot(hm)
   return xm, ym, hm  

def plot_conv(m,xm,ym,hm):
     plt.stem(xm[:,m],markerfmt='+',label='$x[k]$')
     plt.stem(hm,linefmt='g',markerfmt='.',label='$h[n-k]$')
     plt.stem(ym[:m,:].sum(axis=0),markerfmt='s',linefmt='r',label='$y[n]=\sum^{\infty}_{k=-\infty}x[k]h[n-k]$')
     plt.legend()
     plt.show()
     return

xm, ym, hm = convgraf(h,x) 
```
